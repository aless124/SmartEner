```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données depuis le CSV
df = pd.read_csv("CSV/Aless/evolution_de_la_production_d'electricite_en_France_2024-04-04_12-03.csv", sep=';', quotechar='\'', decimal=',', encoding='utf-8')

# Convertir la colonne 'Date' en format de date
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m')

# Définir les types d'énergie comme des séries temporelles
energies = df['Filière'].unique()

# Créer une figure et des sous-graphiques
fig, ax = plt.subplots(figsize=(10, 6))

# Boucler à travers chaque type d'énergie pour tracer
for energy in energies:
    data = df[df['Filière'] == energy]
    ax.plot(data['Date'], data['Valeur (TWh)'], label=energy)

# Ajouter des titres et une légende
ax.set_title('Évolution des différentes sources d\'énergie')
ax.set_xlabel('Date')
ax.set_ylabel('Valeur (TWh)')
ax.legend()

# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import csv

# Charger les données depuis le CSV en échappant les guillemets simples
df = pd.read_csv('CSV\Aless\évolution_de_la_consommation_brute_et_corrigeA9e_2024-04-04_12-03_0.csv', sep=';', decimal=',', quoting=csv.QUOTE_NONE)

# Convertir la colonne 'Date' en format de date
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m')

# Définir les filières comme des séries temporelles
filieres = df['Filière'].unique()

# Créer une figure et des sous-graphiques
fig, ax = plt.subplots(figsize=(10, 6))

# Boucler à travers chaque filière pour tracer
for filiere in filieres:
    data = df[df['Filière'] == filiere]
    ax.plot(data['Date'], data['Valeur (TWh)'], label=filiere)

# Ajouter des titres et une légende
ax.set_title('Évolution des différentes filières énergétiques')
ax.set_xlabel('Date')
ax.set_ylabel('Valeur (TWh)')
ax.legend()

# Afficher le graphique
plt.tight_layout()
plt.show()
```

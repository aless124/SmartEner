```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les donnes depuis le CSV
df = pd.read_csv("CSV/Aless/evolution_de_la_production_d'electricite_en_France_2024-04-04_12-03.csv", sep=';', quotechar='\'', decimal=',', encoding='utf-8')

# Convertir la colonne 'Date' en format de date
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m')

# Dfinir les types d'nergie comme des sries temporelles
energies = df['Filire'].unique()

# Crer une figure et des sous-graphiques
fig, ax = plt.subplots(figsize=(10, 6))

# Boucler  travers chaque type d'nergie pour tracer
for energy in energies:
    data = df[df['Filire'] == energy]
    ax.plot(data['Date'], data['Valeur (TWh)'], label=energy)

# Ajouter des titres et une lgende
ax.set_title('volution des diffrentes sources d\'nergie')
ax.set_xlabel('Date')
ax.set_ylabel('Valeur (TWh)')
ax.legend()

# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import csv

# Charger les donnes depuis le CSV en chappant les guillemets simples
df = pd.read_csv('CSV\Aless\volution_de_la_consommation_brute_et_corrigeA9e_2024-04-04_12-03_0.csv', sep=';', decimal=',', quoting=csv.QUOTE_NONE)

# Convertir la colonne 'Date' en format de date
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m')

# Dfinir les filires comme des sries temporelles
filieres = df['Filire'].unique()

# Crer une figure et des sous-graphiques
fig, ax = plt.subplots(figsize=(10, 6))

# Boucler  travers chaque filire pour tracer
for filiere in filieres:
    data = df[df['Filire'] == filiere]
    ax.plot(data['Date'], data['Valeur (TWh)'], label=filiere)

# Ajouter des titres et une lgende
ax.set_title('volution des diffrentes filires nergtiques')
ax.set_xlabel('Date')
ax.set_ylabel('Valeur (TWh)')
ax.legend()

# Afficher le graphique
plt.tight_layout()
plt.show()
```
